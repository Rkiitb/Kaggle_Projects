# Data Preprocessing

```python
# Let's check the name columns first
df['name'].unique()
```

##### The above are (car names + brand) so we can separate them into two different columns

```python
def brand_name(txt):
    """
    Extracts and returns the brand name from a given car model name.
    Parameters:- txt (str): The input car model name from which the brand name needs to be extracted.
    Returns:- str: The extracted brand name.
    """
    return txt.split()[0]
```

```python
def Car_name(txt):
    """
    Extracts and returns the car name (excluding the brand) from a given car model name.
    Parameters:- txt (str): The input car model name from which the car name needs to be extracted.
    Returns:- str: The extracted car name.

    """
    return ' '.join(txt.split()[1:])
```

```python
df['Car Brand']=df['name'].apply(brand_name)
df['Car Name']=df['name'].apply(Car_name)
df.sample(3)
```

##### Similarly we can do the same for color column. Split both the columns in two columns interior color and exterior color

```python
def interior_color(txt):
    """Extracts and returns the interior color from a given car model name or description.
    Parameters:- txt (str): The input text containing information about the car's interior color.
    Returns:- str: The extracted interior color.
    """
    return txt.split(',')[1]
```

```python
def exterior_color(txt):
    """
    Extracts and returns the exterior color from a given car model name or description.
    Parameters:- txt (str): The input text containing information about the car's exterior color.
    Returns:- str: The extracted exterior color.
    """
    return txt.split(',')[0]
```

```python
df['Interior Color']=df['color'].apply(interior_color).str.replace(' interior','').str.replace(',','')
df['Exterior Color']=df['color'].apply(exterior_color).str.replace(' exterior','').str.replace(',','')
df
```

```python
# To convert the miles column in numerical 
df['miles']=df['miles'].str.replace(' miles','').str.replace(',','').astype(int)
df.sample(2)
```

```python
# remove the $ sign to convert the price value into numerical 
df['price in $']=df['price'].str.replace('$','').str.replace(',','').astype(int)
df.sample(2)
```

```python
df[['accident_reported', 'owner_count']] = df['condition'].str.split(', ', expand=True)
df['owner_count']=df['owner_count'].str.replace(' Owner','').str.strip()
```

```python
df['accident_reported'].unique()
```

```python
#To convert the no of accidents in numerical values
df['accident_reported']=df['accident_reported'].map(
    {'No accidents reported':0,
     '1 accident reported':1,
     '2 accident reported':2,
     '3 accidents reported':3,
     '4 accidents reported':4,
     '5 accidents reported':5
    })
df
```
